```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Read the data file
df = pd.read_excel('airBnb_data_countries_merged.xlsx', sheet_name='data')

# Now you can work with the DataFrame 'df'
df.head()
```

```python
# Get info about the dataset
df.info()
```

```python
# Process to descriptive statistics
df.describe()
```

```python
# Detect null values
null_values = df.isnull().sum()
print(null_values)
```

```python
# Find duplicated values
duplicated_values = df.duplicated().sum()
print(duplicated_values)
```

```python
# Delete columns room_shared and room_private
df = df.drop(['room_shared', 'room_private'], axis=1)
df.info()
```

```python
# Change all columns with booleans to 0 and 1
df = df.replace({True: 1, False: 0})
df.head()
```

```python
# Calculate outliers and remove them using z-score method
df['booking_amount_zscore'] = (df['booking_amount'] - df['booking_amount'].mean()) / df['booking_amount'].std()

# Define threshold for outliers
threshold = 3

# Identify outliers
outliers = df[abs(df['booking_amount_zscore']) > threshold]

# Remove outliers
df = df[abs(df['booking_amount_zscore']) <= threshold]

# Drop the z-score column
df = df.drop('booking_amount_zscore', axis=1)

df.shape
```

```python
# Process label encoding
from sklearn.preprocessing import LabelEncoder
df = df.apply(LabelEncoder().fit_transform)
```

```python
# splitting the data into dependent and independent variables
x = df.drop('booking_amount', axis = 1)
y = df.booking_amount

print(x.shape)
print(y.shape)
```

```python
# making x_train, x_test, y_train, y_test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

x_train = pd.DataFrame(x_train)
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
import time

# instantiating the model
model_lr = LinearRegression()

#Starting the modelling
start_lr = time.time()
model_lr.fit(x_train, y_train)

# predicting the  test set results
y_pred = model_lr.predict(x_test)

#Ending the modelling
end_lr = time.time()

# finding the rmse,r2, mae and execution time of the model
score_one = round(r2_score(y_test, y_pred), 3)
mae_one = round(mean_absolute_error(y_test, y_pred), 3)
rmse_one = round(mean_squared_error(y_test, y_pred), 3)
ex_time_lr = round(end_lr - start_lr, 3)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# instantiating the model
model_gb = GradientBoostingRegressor()

#Starting the modelling
start_gb = time.time()

model_gb.fit(x_train, y_train)

# predicting the test set results
y_pred = model_gb.predict(x_test)

#Ending the modelling
end_gb = time.time()

# finding the rmse,r2, mae and execution time of the model
score_two = round(r2_score(y_test, y_pred), 3)
mae_two = round(mean_absolute_error(y_test, y_pred), 3)
rmse_two = round(mean_squared_error(y_test, y_pred), 3)
ex_time_gb = round(end_gb - start_gb, 3)
```

```python
from sklearn.ensemble import RandomForestRegressor

# instantiating the model
model_rf = RandomForestRegressor()

#Starting the modelling
start_rf = time.time()

model_rf.fit(x_train, y_train)

# predicting the  test set results
y_pred = model_rf.predict(x_test)

#Ending the modelling
end_rf = time.time()

# finding the rmse,r2, mae and execution time of the model
score_three = round(r2_score(y_test, y_pred), 3)
mae_three = round(mean_absolute_error(y_test, y_pred), 3)
rmse_three = round(mean_squared_error(y_test, y_pred), 3)
ex_time_rf = round(end_rf - start_rf, 3)
```

```python
from sklearn.tree import DecisionTreeRegressor

# instantiating the model
model_dt = DecisionTreeRegressor()

#Starting the modelling
start_dt = time.time()

model_dt.fit(x_train, y_train)

# predicting the test set results
y_pred = model_dt.predict(x_test)

#Ending the modelling
end_dt = time.time()

# finding the rmse,r2, mae and execution time of the model
score_fourth = round(r2_score(y_test, y_pred), 3)
mae_fourth = round(mean_absolute_error(y_test, y_pred), 3)
rmse_fourth = round(mean_squared_error(y_test, y_pred), 3)
ex_time_dt = round(end_dt - start_dt, 3)
```

```python
from sklearn.linear_model import Lasso

# instantiating the model
model_lasso = Lasso()

#Starting the modelling
start_lasso = time.time()
model_lasso.fit(x_train, y_train)

# predicting the  test set results
y_pred = model_lasso.predict(x_test)

#Ending the modelling
end_lasso = time.time()

# finding the rmse,r2, mae and execution time of the model
score_five = round(r2_score(y_test, y_pred), 3)
mae_five = round(mean_absolute_error(y_test, y_pred), 3)
rmse_five = round(mean_squared_error(y_test, y_pred), 3)
ex_time_lasso = round(end_lasso - start_lasso, 3)
```

```python
# Cross validation
from sklearn.model_selection import cross_val_score

cvs_rf = cross_val_score(estimator=model_rf , X = x_train, y = y_train , cv =5)
cvs_dt = cross_val_score(estimator=model_dt , X = x_train, y = y_train , cv =5)
cvs_gb = cross_val_score(estimator=model_gb , X = x_train, y = y_train , cv =5)
cvs_lr = cross_val_score(estimator=model_lr , X = x_train, y = y_train , cv =5)
cvs_lasso = cross_val_score(estimator=model_lasso , X = x_train, y = y_train , cv =5)
```

```python
models = pd.DataFrame({
    'Model':["Linear Regression", "XgBoost Regressor","Random Forest Regression", "Decision Tree Regressor", "Lasso Regression"],
    "Score":[score_one, score_two, score_three, score_fourth, score_five],
    "MAE": [mae_one, mae_two, mae_three, mae_fourth, mae_five] ,
    "RMSE": [rmse_one, rmse_two, rmse_three, rmse_fourth, rmse_five] ,
    "EX_TIME": [ex_time_lr, ex_time_gb, ex_time_rf, ex_time_dt, ex_time_lasso] ,
    "CROSS": [round(cvs_lr.mean()*100 , 2), round(cvs_gb.mean()*100 , 2),
              round(cvs_rf.mean()*100 , 2), round(cvs_dt.mean()*100 , 2),
              round(cvs_lasso.mean()*100 , 2)] ,
})
models.sort_values(by='Score',ascending=False)
```

```python
# Plot the model r2 score
plt.figure(figsize=(10, 6))
sns.boxplot(x='Score', y='Model', data=models, dodge=False, hue="Model")
plt.title('Model Comparison - R2 Score')
plt.xlabel('R2 Score')
plt.ylabel('Models')
plt.yticks(rotation=45)
plt.show()
```

```python
# Plot a residual plot to visualize the model's performance

# Calculate residuals
residuals = y_test - model_rf.predict(x_test)

# Create a residual plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, residuals)
plt.xlabel('Actual Values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()

# Visualized Predicted vs Actual plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, model_rf.predict(x_test))
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Predicted vs Actual Values')
plt.show()
```

```python
# Residual Plot:
# - If the points are randomly scattered around the horizontal line at y=0, it suggests that the model's errors are random and have constant variance. This is a good sign
# - If there's a pattern in the residuals (e.g., a curve or a funnel shape), it indicates that the model might not be capturing all the relationships in the data

# Predicted vs Actual Plot:
# - A strong diagonal line indicates a good fit
# - Deviations from the diagonal line suggest areas where the model is under- or over-predicting

# To draw specific conclusions, you need to carefully examine the plots you generated
# Do the residuals show any patterns?
# Does the predicted vs actual plot have a strong diagonal line?

# Based on your observations, you can suggest:
# - If the model is a good fit for the data
# - If there are any areas where the model could be improved
# - If there are any potential issues with the data or the model
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a GridSearchCV object
grid_search_rf = GridSearchCV(estimator=RandomForestRegressor(), param_grid=param_grid_rf, cv=5, scoring='r2')

# Fit the grid search to the data
grid_search_rf.fit(x_train, y_train)

# Get the best parameters
best_params_rf = grid_search_rf.best_params_

# Train the model with the best parameters
best_model_rf = RandomForestRegressor(**best_params_rf)
best_model_rf.fit(x_train, y_train)

# Predict on the test set
y_pred_rf = best_model_rf.predict(x_test)

# Evaluate the model
score_rf = round(r2_score(y_test, y_pred_rf), 3)
mae_rf = round(mean_absolute_error(y_test, y_pred_rf), 3)
rmse_rf = round(np.sqrt(mean_squared_error(y_test, y_pred_rf)), 3)

# Print the results
print(f"Optimized Random Forest - R2 Score: {score_rf}")
print(f"Optimized Random Forest - MAE: {mae_rf}")
print(f"Optimized Random Forest - RMSE: {rmse_rf}")
```

```python
# Define the parameter grid for Gradient Boosting
param_grid_gb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a GridSearchCV object
grid_search_gb = GridSearchCV(estimator=GradientBoostingRegressor(), param_grid=param_grid_gb, cv=5, scoring='r2')

# Fit the grid search to the data
grid_search_gb.fit(x_train, y_train)

# Get the best parameters
best_params_gb = grid_search_gb.best_params_

# Train the model with the best parameters
best_model_gb = GradientBoostingRegressor(**best_params_gb)
best_model_gb.fit(x_train, y_train)

# Predict on the test set
y_pred_gb = best_model_gb.predict(x_test)

# Evaluate the model
score_gb = round(r2_score(y_test, y_pred_gb), 3)
mae_gb = round(mean_absolute_error(y_test, y_pred_gb), 3)
rmse_gb = round(np.sqrt(mean_squared_error(y_test, y_pred_rf)), 3)

# Print the results
print(f"Optimized Gradient Boosting - R2 Score: {score_gb}")
print(f"Optimized Gradient Boosting - MAE: {mae_gb}")
print(f"Optimized Gradient Boosting - RMSE: {rmse_gb}")
```
